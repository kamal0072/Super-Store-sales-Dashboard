```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load, 
df = pd.read_csv("../csvFiles/Sample_Superstore.csv", encoding="latin-1")
# view,
df.head(5)
#  understand, 
df.shape
# columns name,
df.columns
# data types,
df.dtypes
#  summery stats,  
df.describe()
```

```python
# missing values
df.isnull().sum()
# duplicates
df.duplicated()
df[df.duplicated()]
# remove duplicates
df.drop_duplicates()
# df
```

```python
# Convert Date
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date'].dtype
```

```python
# Create Month, Year Column
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
df
```

```python
# Business KPIs
# Total Sales, Total Profits, Total Orders, Total Customers , 
total_sales = df['Sales'].sum()
total_sales
total_profits = df['Profit'].sum()
total_profits
total_orders = df['Order ID'].nunique()
total_orders
total_customers = df['Customer ID'].nunique()
total_customers

```

```python
# Average Order Value
avg_order_value =  df['Sales'].mean()
avg_order_value
```

```python
monthly = df.groupby("Month")['Sales'].sum().reset_index()
monthly
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=monthly,
    x="Month",
    y='Sales',
    markers="o"
)
plt.xticks(rotation=90)
plt.show()
```

```python
category = df.groupby('Category')['Sales'].sum().reset_index()
category
sns.barplot(
    data=category,
    x="Category",
    y="Sales",
)
plt.show()
```
